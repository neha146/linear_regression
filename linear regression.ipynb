```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def linear_regression(x_train, y_train):
    # Add a column of ones to x_train for the intercept term
    ones = np.ones((x_train.shape[0], 1))
    x_train = np.concatenate((ones, x_train), axis=1)
    
    # Calculate the parameters using the closed-form solution
    x_transpose = np.transpose(x_train)
    x_transpose_dot_x = np.dot(x_transpose, x_train)
    inverse = np.linalg.inv(x_transpose_dot_x)
    x_transpose_dot_y = np.dot(x_transpose, y_train)
    theta = np.dot(inverse, x_transpose_dot_y)
    
    return theta


# Data preprocessing and visualization
data = pd.read_csv('Medical Price Dataset.csv')  # Replace 'your_data.csv' with your dataset file

# Preprocessing (if required)
# You can perform data preprocessing steps such as normalization, handling missing values, etc., here.

# Splitting features and target
x_train = data[['age','sex','region','children','smoker', 'bmi']]  # Replace 'feature1' and 'feature2' with your actual feature column names
y_train = data['charges']  # Replace 'target' with your actual target column name

# Visualization (if required)
# You can visualize your data using scatter plots, histograms, etc., to gain insights.

# Scatter plot
plt.scatter(x_train['age'], y_train)
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Scatter plot')
plt.show()

plt.scatter(x_train['bmi'], y_train)
plt.xlabel('BMI')
plt.ylabel('Charges')
plt.title('Scatter plot')
plt.show()

plt.scatter(x_train['region'], y_train)
plt.xlabel('region')
plt.ylabel('Charges')
plt.title('Scatter plot')
plt.show()

plt.scatter(x_train['sex'], y_train)
plt.xlabel('Sex')
plt.ylabel('Charges')
plt.title('Scatter plot')
plt.show()

plt.scatter(x_train['smoker'], y_train)
plt.xlabel('smoker')
plt.ylabel('Charges')
plt.title('Scatter plot')
plt.show()



# Histogram
plt.hist(x_train['children'], bins=10)
plt.xlabel('children')
plt.ylabel('Charges')
plt.title('Histogram')
plt.show()
```
